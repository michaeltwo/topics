```python
import os
import pandas as pd
import chardet

def process_files(root_folder, output_file):
    # 用于存储所有处理后的数据
    all_data = []
    cols_to_keep = ["url", "date", "full text", "impressions", "twitter followers","gender","sentiment","region"]

    # 遍历文件夹及子文件夹
    for dirpath, _, filenames in os.walk(root_folder):
        for file in filenames:
            file_path = os.path.join(dirpath, file)
            # 检查文件后缀
            if file.endswith('.csv'):
                try:
                    with open(file_path, 'rb') as file:
                        result = chardet.detect(file.read())
                        # print(result)  # 输出编码格式
                    # 读取CSV文件，跳过前6行
                    data = pd.read_csv(file_path, skiprows=6,encoding=result['encoding'])
                    data.columns = data.columns.str.lower()
                    columns_to_keep = [col.lower() for col in cols_to_keep]
                    data = data[columns_to_keep]
                    all_data.append(data)
                except Exception as e:
                    print(f"无法处理CSV文件 {file_path}: {e}")
            elif file.endswith('.xls') or file.endswith('.xlsx'):
                try:
                    # with open(file_path, 'rb') as file:
                    #     result = chardet.detect(file.read())
                    # 读取Excel文件，跳过前6行
                    data = pd.read_excel(file_path, skiprows=6)
                    data.columns = data.columns.str.lower()
                    columns_to_keep = [col.lower() for col in cols_to_keep]
                    data = data[columns_to_keep]
                    all_data.append(data)
                except Exception as e:
                    print(f"无法处理Excel文件 {file_path}: {e}")

    # 将所有数据拼接成一个文件
    if all_data:
        combined_data = pd.concat(all_data, ignore_index=True)
        # 导出到指定的文件
        if output_file.endswith('.csv'):
            combined_data.to_csv(output_file, index=False)
        elif output_file.endswith('.xlsx'):
            combined_data.to_excel(output_file, index=False, engine='openpyxl')
        print(f"所有文件已合并到 {output_file}")
    else:
        print("未找到可处理的文件。")
```

```python
# 使用示例
input_directory = "../root/"  # 替换为实际的输入目录路径
output_filepath = "../merged_file.csv"  # 替换为实际的输出CSV文件路径
process_files(input_directory, output_filepath)
```
